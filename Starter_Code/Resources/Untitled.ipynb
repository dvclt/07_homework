```python
# Import the required libraries and dependencies
import pandas as pd
import hvplot.pandas
from pathlib import Path
```

```python
# Using the read_csv function and Path module, create a DataFrame 
# by importing the sfo_neighborhoods_census_data.csv file from the Resources folder
sfo_data_df = pd.read_csv(Path('../Resources/sfo_neighborhoods_census_data.csv'))
# Review the first and last five rows of the DataFrame
display(sfo_data_df.head())
display(sfo_data_df.tail())
```

```python
sfo_years = sfo_data_df[['year','housing_units']]
sfo_years
```

```python
plot_yrs = sfo_years.groupby(['year']).mean().hvplot.bar(
    xlabel='Year',
    ylabel='Housing Units',
    title='Housing Units in San Francisco from 2010 to 2016',
    ylim=[365000,385000]
).opts(yformatter='%.0f', )
plot_yrs
```

```python
sfo_sale_rent=sfo_data_df.groupby(['year']).mean()
sfo_sale_rent
```

```python
plot_sales_rent = sfo_sale_rent[['sale_price_sqr_foot','gross_rent']].hvplot.line(
    xlabel='Year',
    ylabel='Gross Rent / Sale Price per Square Foot',
    title='Sale Price per Square Foot and Average Gross Rent - 2010-2016 - San Francisco',
).opts(yformatter='%.0f', )
plot_sales_rent 
```

```python
sfo_data_df['neighborhood'].value_counts()
```

```python
sfo_hood =sfo_data_df[['year','neighborhood','sale_price_sqr_foot','gross_rent']].groupby(['year','neighborhood']).mean()
sfo_hood
```

```python
plot_hood = sfo_hood.hvplot(
    xlabel='Year',
    ylabel='Gross Rent / Sale Price per Square Foot',
    title='Sale Price per Square Foot and Average Gross Rent - 2010-2016 - San Francisco',
    groupby='neighborhood'
).opts(yformatter='%.0f', )
plot_hood
```

```python
neighborhood_locations_df = pd.read_csv(Path('../Resources/neighborhoods_coordinates.csv'),index_col='Neighborhood')
neighborhood_locations_df.head()
```

```python
all_neighborhood_info_df= sfo_data_df.groupby(['neighborhood']).mean().drop('year', axis=1)
all_neighborhood_info_df.head()
```

```python
all_neighborhoods_df = pd.concat(
    [all_neighborhood_info_df,neighborhood_locations_df],
    axis=1).dropna()
all_neighborhoods_df.head()
```

```python
plot_map= all_neighborhoods_df.hvplot.points(
    'Lon',
    'Lat',
    geo=True,
    scale=0.8,
    tiles='OSM',
    color='gross_rent',
    size='sale_price_sqr_foot',
    size_max=25,
    zoom=11,
    frame_width=700,
    frame_height=500
)

plot_map
```
